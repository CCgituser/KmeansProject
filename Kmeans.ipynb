```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv(r"C:\Users\Craigs\Downloads\bank notes.csv") # read in the data
mean_values = data.mean()# Calculate statistical measures
std_dev_values = data.std()
print("Mean values:")
print(mean_values)
print("\nStandard deviation values:")
print(std_dev_values)
plt.scatter(data['V1'], data['V2'])# Plot the scatter plot
plt.xlabel('V1')
plt.ylabel('V2')
plt.title('Scatter Plot of V1 vs V2')
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

# Step1 and step 2. Running k means on the bank note data and visualising
# Step 3 the k-means algorithm ran from k=3 to k=9
# Step 4 By comparing the results the algorithm appears to be stable as k increases
# Step 5 as k increases from 3 to 9 the fit becomes "tighter" and fits the data very well

```python

bank_notes = np.column_stack((data['V1'], data['V2']))
km_res = KMeans(n_clusters=9).fit(bank_notes) 
clusters = km_res.cluster_centers_
plt.scatter(data['V1'], data['V2'])
plt.scatter(clusters[:,0], clusters[:,1])
plt.xlabel('V1')
plt.ylabel('V2')
plt.title('Forged banknote detection algorithm')
plt.grid(True)
plt.show() 
```
